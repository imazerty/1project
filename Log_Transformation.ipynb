```
%matplotlib inline
import imageio
import numpy as np
import matplotlib.pyplot as plt
pic = imageio.imread('yft_neg.jpg')
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray1 = gray(pic)

pic2 = imageio.imread('Yellowfin+Tuna+Loin+Precooked+-+Yellowfin+Tuna.jpg')
gray2 = gray(pic2)
'''
log transform
-> s = c*log(1+r)
So, we calculate constant c to estimate s
-> c = (L-1)/log(1+|I_max|)
'''
max_ = np.max(gray1)
max_2 = np.max(gray2)
def log_transform(gray1, max_):
    return (255/np.log(1+max_)) * np.log(1+gray1)
plt.figure(figsize = (5,5))
plt.imshow(log_transform(gray1, max_), cmap = plt.get_cmap(name = 'gray'))
plt.axis('off');
```
